# Linear Regression



Linear Regression tries to fit a straight line:

$$
y = mx + b
$$

- $m$ → slope (weight)  
- $b$ → intercept (bias)  

**Goal:** find $m$ and $b$ such that predictions are as close as possible to actual values.


```python
import numpy as np #Importing Numpy for calculation
```

```python
import matplotlib.pyplot as plt #Importing matplotlib for visualization of our line
```

```python
#Sample data
X = np.array([1,2,3,4,5])
Y = np.array([1,2,3,4,5])

m=0
b=0
```

```python
learning_rate = 0.01
epochs =50
n = len(X)
```

```python
plt.ion()
```

## Loss Function (RMSE)

Instead of just eyeballing the line, we need a loss function. Here we use **Root Mean Squared Error (RMSE):**

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2}
$$

- $y_i$ → actual value  
- $\hat{y}_i$ → predicted value  
- $n$ → number of data points  

RMSE tells us **“on average, how far off our predictions are.”**

## Gradients (Core Idea)

We update parameters using gradient descent:

$$
m = m - \alpha \cdot \frac{\partial MSE}{\partial m}
$$

$$
b = b - \alpha \cdot \frac{\partial MSE}{\partial b}
$$

Where:

- $\alpha$ = learning rate  

---

## Derivatives

$$
\frac{\partial MSE}{\partial m} = -\frac{2}{n} \sum x_i (y_i - \hat{y}_i)
$$

$$
\frac{\partial MSE}{\partial b} = -\frac{2}{n} \sum (y_i - \hat{y}_i)
$$

The rest is just for visualiztion

```python

for epoch in range(epochs):
    y_pred = m * X + b

    # Computing gradients
    dm = (-2/n)*np.sum(X*(Y -y_pred))
    db = (-2/n)*np.sum(Y- y_pred)

    # Updating the  parameters
    m -= learning_rate*dm
    b -= learning_rate*db

    # Plot every few steps
    if epoch % 5 == 0:
        plt.clf()  # Clear previous plot

        # Scatter plot (actual data)
        plt.scatter(X, Y, label="Data Points")

        # Regression line for visuals
        plt.plot(X, y_pred, label=f"Epoch {epoch}", linestyle='--')

        plt.title("Linear Regression Training (Gradient Descent)")
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.legend()

        plt.pause(0.3)  # Pause to see updates

plt.ioff()
plt.show()

# Final RMSE
rmse = np.sqrt(np.mean((Y - y_pred)**2))
print("Final m:", m)
print("Final b:", b)
print("Final RMSE:", rmse)
```
